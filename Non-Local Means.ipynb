```python
# Imprting the Libraries
import cv2
import numpy as np
from skimage.restoration import denoise_nl_means, estimate_sigma
from google.colab.patches import cv2_imshow # This is for displaying images in Google Colab
```

```python
# Reading the Image
image=cv2.imread("/content/image.jpg")
```

```python
# Resizing the Image
image=cv2.resize(image, (400, 400))
```

```python
# Creating the Noisy Image
image_noisy=np.random.poisson(image)
```

```python
# Estimating the Sigma
sigma = np.mean(estimate_sigma(image_noisy, multichannel=True))
```

```python
# Denoising the Image
denoising = denoise_nl_means(image_noisy, h= sigma, fast_mode=True,
                               patch_size=5, patch_distance=5, multichannel=False)
```

```python
# Displaying Function
def cv2disp(name, image):
  cv2_imshow(image)
```

```python
# Displaying the Main Image, Noisy Image, Kernel, and Output Image
cv2disp('Main', image)
cv2disp('Noisy', image_noisy)
cv2disp('ouput', denoising)
```
